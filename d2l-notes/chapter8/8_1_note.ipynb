### Upsample Image

`d2l.FashionMNIST` does:

```python
trans = transforms.Compose([
    transforms.Resize(resize),
    transforms.ToTensor()
])
```

So when you call:

```python
d2l.FashionMNIST(batch_size=128, resize=(224, 224))
```

each original `28 x 28` Fashion-MNIST PIL image is first resized to `224 x 224`, then converted to a tensor. 

It is **not padding around the border**. It is not like putting the old `28 x 28` image in the middle of a larger blank canvas. Instead, it **stretches/resamples the whole image over the full `224 x 224` grid**.

Because `transforms.Resize` defaults to **bilinear interpolation** for PIL images, each new pixel is usually a weighted average of nearby old pixels. The official torchvision docs say `Resize` rescales the image and defaults to bilinear interpolation

A 1D intuition:

Original pixels:

```text
[10, 20, 30]
```

If we resize to more pixels, new locations are mapped back into the old coordinate system. A new pixel that lands halfway between `10` and `20` becomes:

```text
0.5 * 10 + 0.5 * 20 = 15
```

A new pixel closer to `20`, say 25% from `20` to `30`, becomes:

```text
0.75 * 20 + 0.25 * 30 = 22.5
```

For 2D images, bilinear interpolation does this in both directions. A new pixel is computed from the four neighboring old pixels:

```text
top-left, top-right,
bottom-left, bottom-right
```

So conceptually:

```text
new pixel = weighted average of 4 nearby old pixels
```

For Fashion-MNIST `28 -> 224`, the scaling factor is `8`. The new pixels are distributed **inside the whole old image area**, not just added around the edge. The original image is smoothly enlarged; no new information is created.

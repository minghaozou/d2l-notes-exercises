### $\binom{k-1+d}{k-1}$

A monomial of total degree $d$ in $k$ variables $x_1, \ldots, x_k$ looks like
$$
x_1^{a_1} x_2^{a_2} \cdots x_k^{a_k}
$$
where the exponents are nonnegative integers and sum to $d$ :
$$
a_1+\cdots+a_k=d, \quad a_i \in \mathbb{Z}_{\geq 0} .
$$
So the problem is: how many $k$-tuples $\left(a_1, \ldots, a_k\right)$ of nonnegative integers sum to $d$?

This is the classic stars and bars counting argument:
* Imagine $d$ identical "stars" in a row (these are the degree units).
* You want to split them into $k$ groups, giving $a_1$ stars to variable $1, \ldots, a_k$ stars to variable $k$.
* To create $k$ groups, you place $k-1$ "bars" among the stars (including possibly at the ends, which allows some $a_i=0$ ).

There are $d$ stars and $k-1$ bars total, so $d+k-1$ positions. Choosing which $k-1$ positions are bars determines the split uniquely. Hence the count is
$$
\binom{d+k-1}{k-1}=\binom{d+k-1}{d} .
$$
That's exactly $\binom{k-1+d}{k-1}$.

Start from
$$
\binom{d+k-1}{k-1}=\frac{(d+k-1)!}{(k-1)!d!}=\frac{(d+1)(d+2) \cdots(d+k-1)}{(k-1)!} .
$$
The numerator is a product of $k-1$ terms each $\sim d$ when $d$ is large, so the leading growth is polynomial of degree $k-1$ :
$$
\binom{d+k-1}{k-1} \sim \frac{d^{k-1}}{(k-1)!} \quad \text { as } d \rightarrow \infty .
$$
So the complexity in terms of $d$ is:
$$
\binom{d+k-1}{k-1}=\Theta\left(d^{k-1}\right) \quad \text { for fixed } k .
$$
(And more precisely, it's a degree-( $k-1$ ) polynomial in $d$ divided by ( $k-1$ )!, with leading term $d^{k-1} /(k-1)!$.)

### Parameter groups inside `torch.optim.SGD`

Normally, we might see:

```
optimizer = torch.optim.SGD(model.parameters(), lr=0.01)
```

This means:
* Apply SGD to all parameters with the same hyperparameters

But PyTorch also allows something more flexible:
```
optimizer = torch.optim.SGD(param_groups, lr=...)
```
where `param_groups` is a list of dictionaries.

Each dictionary defines:
* which parameters it controls
* optional hyperparameters for those parameters

This is called parameter groups.

````python
[
    {'params': self.net.weight, 'weight_decay': self.wd},
    {'params': self.net.bias}
]
````

This is a list with two dictionaries so we are defining two parameter groups.

````python
{'params': self.net.weight, 'weight_decay': self.wd}
````

This means:
* Optimize `self.net.weight`
* Apply L2 regularization (weight_decay) with strength `self.wd`

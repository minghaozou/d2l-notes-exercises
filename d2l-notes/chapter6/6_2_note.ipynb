### Lazy Initialization

What is nn.LazyLinear?

When you write:
```
shared = nn.LazyLinear(8)
```
you are saying:

“I want a linear layer with output dimension 8, but I don’t yet know the input dimension — figure it out later.”

So internally:

* `weight` and `bias` are not initialized yet
* Their shapes are unknown
* They are basically placeholders

When are parameters created?

They are initialized during the first forward pass, when PyTorch sees actual data:
```
net(X)
```
At that moment:

* PyTorch looks at `X.shape`
* Infers input dimension (say `d`)
* Creates `weight` with shape `(8, d)`
* Initializes parameters (e.g., with default init)

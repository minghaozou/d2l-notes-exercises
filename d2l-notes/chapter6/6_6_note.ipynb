```python
import torch
from torch import nn
from torch.nn import functional as F
```

```python
class MLP(nn.Module):
    def __init__(self):
        super().__init__()
        self.hidden = nn.LazyLinear(256)
        self.output = nn.LazyLinear(10)

    def forward(self, x):
        return self.output(F.relu(self.hidden(x)))

net = MLP()
X = torch.randn(size=(2, 20))
Y = net(X)
```

```python
torch.save(net.state_dict(), 'mlp.params')
```

```python
clone = MLP()
clone.load_state_dict(torch.load('mlp.params'))
clone.eval()
```

```python
print(clone.hidden.weight.shape)   
print(clone.output.weight.shape)   
print(clone)
```

Loading parameters does not necessarily mean the lazy layer’s descriptive attributes have already been fully materialized.
The tensors can be loaded first, and the layer’s lazy metadata can catch up on the first forward pass.

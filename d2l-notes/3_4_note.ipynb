### Optimizer

#### Why do `params` have a `.grad` attribute?

```python
self.w = torch.normal(0, sigma, (num_inputs, 1), requires_grad=True)
self.b = torch.zeros(1, requires_grad=True)
```

Because we set

```python
requires_grad=True
```

these tensors are now leaf tensors in the computation graph.

That means:

* PyTorch tracks operations involving them

* After calling .backward(), PyTorch stores gradients in:

```python
param.grad
```

They are torch.Tensor objects, and tensors with `requires_grad=True` get a `.grad` attribute.

#### Why does `param.grad.zero_()` exist?

Because `param.grad` is itself a tensor.

In PyTorch:

* `zero_()` means in-place zeroing

* The underscore `_` always means in-place operation

So:

```python
param.grad.zero_()
```

means:

* Set all gradient entries to 0, without allocating new memory.

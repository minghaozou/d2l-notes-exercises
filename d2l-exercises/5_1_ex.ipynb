### Exercise 1

Show that adding layers to a linear deep network, i.e., a network without nonlinearity $\sigma$ can never increase the expressive power of the network. Give an example where it actively reduces it.

In a linear network without activation functions, composing layers gives another affine map. For example,

$$
\mathbf{O}=\left(\mathbf{X} \mathbf{W}^{(1)}+\mathbf{b}^{(1)}\right) \mathbf{W}^{(2)}+\mathbf{b}^{(2)}=\mathbf{X}\left(\mathbf{W}^{(1)} \mathbf{W}^{(2)}\right)+\left(\mathbf{b}^{(1)} \mathbf{W}^{(2)}+\mathbf{b}^{(2)}\right),
$$

so the whole network is equivalent to a single affine layer. Hence adding linear layers can never increase expressive power.

It can, however, reduce expressive power if an added hidden layer is narrower than the input/output dimension. For instance, consider maps $\mathbb{R}^2 \rightarrow \mathbb{R}^2$. A single linear layer can represent any $2 \times 2$ matrix, including the identity matrix. But a two-layer linear network $\mathbb{R}^2 \rightarrow \mathbb{R}^1 \rightarrow \mathbb{R}^2$ can only produce matrices of rank at most 1, since the effective matrix factors through a 1-dimensional hidden space. Therefore it cannot represent the identity map, whose rank is 2 . Thus adding layers can actually reduce expressive power.

### Exercise 2

Compute the derivative of the pReLU activation function.

$$
\frac{d}{d x} \operatorname{pReLU}(x)= \begin{cases}1 & x>0 \\ \alpha & x<0\end{cases}
$$

and the derivative is undefined at $x=0$ (though in practice a subgradient is used).

### Exercise 3

Compute the derivative of the Swish activation function $x \operatorname{sigmoid}(\beta x)$.

$$
\frac{d}{d x}(x \sigma(\beta x))=\sigma(\beta x)+\beta x \sigma(\beta x)(1-\sigma(\beta x))
$$

or equivalently

$$
\sigma(\beta x)[1+\beta x(1-\sigma(\beta x))]
$$

The derivative is basically

$$
\text { sigmoid }+x \times \text { (sigmoid gradient) }
$$

which is why Swish has smooth gradients, one of the reasons it works well in deep networks.

### Exercise 4

Show that an MLP using only ReLU (or pReLU) constructs a continuous piecewise linear function.

A ReLU network does not create curves in the usual smooth sense. Instead, it cuts the input space into many regions, and on each region the network acts like a different affine function. The “bends” happen only at boundaries where some neuron changes from inactive to active.

So the network is like a big patchwork of linear pieces glued together continuously.

### Exercise 1

Show that adding layers to a linear deep network, i.e., a network without nonlinearity $\sigma$ can never increase the expressive power of the network. Give an example where it actively reduces it.

In a linear network without activation functions, composing layers gives another affine map. For example,

$$
\mathbf{O}=\left(\mathbf{X} \mathbf{W}^{(1)}+\mathbf{b}^{(1)}\right) \mathbf{W}^{(2)}+\mathbf{b}^{(2)}=\mathbf{X}\left(\mathbf{W}^{(1)} \mathbf{W}^{(2)}\right)+\left(\mathbf{b}^{(1)} \mathbf{W}^{(2)}+\mathbf{b}^{(2)}\right),
$$

so the whole network is equivalent to a single affine layer. Hence adding linear layers can never increase expressive power.

It can, however, reduce expressive power if an added hidden layer is narrower than the input/output dimension. For instance, consider maps $\mathbb{R}^2 \rightarrow \mathbb{R}^2$. A single linear layer can represent any $2 \times 2$ matrix, including the identity matrix. But a two-layer linear network $\mathbb{R}^2 \rightarrow \mathbb{R}^1 \rightarrow \mathbb{R}^2$ can only produce matrices of rank at most 1, since the effective matrix factors through a 1-dimensional hidden space. Therefore it cannot represent the identity map, whose rank is 2 . Thus adding layers can actually reduce expressive power.

### Exercise 3

Compute the derivative of the Swish activation function $x \operatorname{sigmoid}(\beta x)$.

$$
\frac{d}{d x}(x \sigma(\beta x))=\sigma(\beta x)+\beta x \sigma(\beta x)(1-\sigma(\beta x))
$$

or equivalently

$$
\sigma(\beta x)[1+\beta x(1-\sigma(\beta x))]
$$

The derivative is basically

$$
\text { sigmoid }+x \times \text { (sigmoid gradient) }
$$

which is why Swish has smooth gradients, one of the reasons it works well in deep networks.

### Exercise 5

Sigmoid and tanh are very similar.
1. Show that $\tanh (x)+1=2 \operatorname{sigmoid}(2 x)$.
2. Prove that the function classes parametrized by both nonlinearities are identical. Hint: affine layers have bias terms, too.

#### 5.1

Start from the definitions:

$$
\tanh (x)=\frac{e^x-e^{-x}}{e^x+e^{-x}}
$$

and

$$
\operatorname{sigmoid}(z)=\frac{1}{1+e^{-z}} .
$$


Now compute $\tanh (x)+1$ :

$$
\tanh (x)+1=\frac{e^x-e^{-x}}{e^x+e^{-x}}+1=\frac{e^x-e^{-x}+e^x+e^{-x}}{e^x+e^{-x}}=\frac{2 e^x}{e^x+e^{-x}} .
$$


Multiply numerator and denominator by $e^x$ :

$$
\frac{2 e^x}{e^x+e^{-x}}=\frac{2 e^{2 x}}{e^{2 x}+1}=\frac{2}{1+e^{-2 x}}
$$


But this is exactly

$$
2 \text { sigmoid( } 2 x \text { ). }
$$


So we have shown

$$
\tanh (x)+1=2 \operatorname{sigmoid}(2 x)
$$


Equivalently,

$$
\tanh (x)=2 \operatorname{sigmoid}(2 x)-1 .
$$

#### 5.2

From part 1,

$$
\tanh (x)=2 \operatorname{sigmoid}(2 x)-1 .
$$


Suppose a hidden unit in a tanh network is

$$
h=\tanh (a) \quad \text { where } a=W x+b .
$$


Then

$$
h=2 \operatorname{sigmoid}(2 a)-1=2 \operatorname{sigmoid}(2 W x+2 b)-1 .
$$


So a tanh unit can be implemented by:
- scaling the pre-activation by 2,
- applying sigmoid,
- then scaling by 2 and shifting by -1 .

Since linear layers have weights and biases, these extra scalings and shifts can be absorbed into adjacent affine layers.

So anything representable with tanh is representable with sigmoid.

The crucial issue is the constant shifts -1 and +1 .
For example,

$$
\tanh (x)=2 \operatorname{sigmoid}(2 x)-1 .
$$


That -1 must be implemented somewhere. If affine layers had no bias terms, handling such constant shifts would be harder. But since affine layers do include biases, these constant offsets are easy to absorb.

That is exactly why the hint says: affine layers have bias terms, too.

### Exercise 6

Assume that we have a nonlinearity that applies to one minibatch at a time, such as the batch normalization (loffe and Szegedy, 2015). What kinds of problems do you expect this to cause?

Because the nonlinearity depends on minibatch statistics, the output for a single example depends on the other examples in the batch. This causes several issues. First, the prediction for a given input is not deterministic and may change depending on which other samples appear in the minibatch. Second, the method becomes sensitive to batch size: small minibatches produce noisy estimates of the mean and variance and can destabilize training. Finally, the dependence between samples complicates optimization and makes it difficult to apply the model in settings with very small batches, streaming data, or single-example inference.

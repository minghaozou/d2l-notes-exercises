```python
import torch
from torch import nn
```

### Exercise 1

Design a layer that takes an input and computes a tensor reduction, i.e., it returns $y_k=\sum_{i, j} W_{i j k} x_i x_j$.

```python
class QuadraticLayer(nn.Module):
    def __init__(self, in_units, out_units):
        super().__init__()
        self.in_units = in_units
        self.out_units = out_units
        
        # W shape: (in_units, in_units, out_units)
        self.weight = nn.Parameter(
            torch.randn(in_units, in_units, out_units)
        )

    def forward(self, X):
        # X shape: (batch_size, in_units)
        # The output shape should be (batch_size, out_units)

        # firstly construct the outer product of X with itself
        # outer shape: (batch_size, in_units, in_units)
        outer = X.unsqueeze(2) * X.unsqueeze(1) # X.unsqueeze(2) shape: (batch_size, in_units, 1), X.unsqueeze(1) shape: (batch_size, 1, in_units), so the result of the multiplication has shape (batch_size, in_units, in_units)

        # then we can use einsum to compute the output
        # outer:  (b, i, j)
        # weight: (i, j, k)
        # result: (b, k)
        Y = torch.einsum('bij,ijk->bk', outer, self.weight)
        return Y
```

```python
layer = QuadraticLayer(4, 3)
X = torch.rand(2, 4)
Y = layer(X)

print(X.shape)   # (2, 4)
print(Y.shape)   # (2, 3)
print(Y)
```

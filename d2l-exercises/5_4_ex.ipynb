### Exercise 1

Look up analytic bounds on the eigenvalues of the product of two matrices. What does this tell you about ensuring that gradients are well conditioned?

For the product of two matrices $A$ and $B$, a few standard analytic bounds are especially important:

$$
\rho(A B) \leq\|A B\| \leq\|A\|\|B\|,
$$

so every eigenvalue of $A B$ is bounded in magnitude by any submultiplicative matrix norm, and in particular by $\|A\|\|B\|$.

For the operator 2-norm, this gives

$$
\sigma_{\max }(A B) \leq \sigma_{\max }(A) \sigma_{\max }(B) .
$$


This is the cleanest upper bound on expansion. It follows from submultiplicativity of induced norms.

If $A$ and $B$ are invertible, then using

$$
(A B)^{-1}=B^{-1} A^{-1}
$$

and norm submultiplicativity,

$$
\sigma_{\min }(A B)=\frac{1}{\left\|(A B)^{-1}\right\|_2} \geq \frac{1}{\left\|B^{-1}\right\|_2\left\|A^{-1}\right\|_2}=\sigma_{\min }(A) \sigma_{\min }(B) .
$$


So the smallest singular value of a product cannot be worse than the product of the smallest singular values. This is a direct consequence of the norm bound above.


Backprop multiplies many Jacobians together. Even if each layer is only mildly ill-conditioned, the whole gradient map can become very ill-conditioned because condition numbers multiply. If some layers have $\sigma_{\max } < 1$, gradients can vanish; if some have $\sigma_{\min } > 1$, gradients can explode; if different singular values vary a lot, gradients become anisotropic and optimization gets unstable.

So "well-conditioned gradients" means trying to keep every layer Jacobian close to an isometry:

$$
\sigma_{\max }\left(J_{\ell}\right) \approx 1, \quad \sigma_{\min }\left(J_{\ell}\right) \approx 1,
$$

That is the real lesson of these matrix-product bounds:
- It is not enough that eigenvalues merely be bounded.
- What matters most is controlling the entire singular value spectrum of each layer.
- Small per-layer distortions compound multiplicatively through depth.

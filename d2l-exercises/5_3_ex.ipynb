### Exercise 5

Assume that the computational graph is too large for your GPU.
1. Can you partition it over more than one GPU?
2. What are the advantages and disadvantages over training on a smaller minibatch?

#### 5.1

Yes. This is model parallelism, where different parts of the neural network run on different GPUs.

#### 5.2

| Method | Advantages | Disadvantages |
| :--- | :--- | :--- |
| Model parallelism | train very large models; keep large batches | communication overhead; complex |
| Smaller minibatch | simple; no communication cost | noisier gradients; worse GPU utilization; cannot solve huge model memory |

Between GPUs we must send tensors:

Forward pass:

$$
h \rightarrow \text { next GPU }
$$


Backward pass:

$$
\frac{\partial L}{\partial h} \rightarrow \text { previous GPU }
$$


GPU communication (PCle/NVLink) is much slower than GPU memory access.
Thus training may slow down.

GPUs work best with large matrix operations.
Very small batches reduce efficiency.

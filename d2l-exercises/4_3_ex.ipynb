### Exercise 1

Denote by $L_{\mathrm{v}}$ the validation loss, and let $L_{\mathrm{v}}^{\mathrm{q}}$ be its quick and dirty estimate computed by the loss function averaging in this section. Lastly, denote by $l_{\mathrm{v}}^{\mathrm{b}}$ the loss on the last minibatch. Express $L_{\mathrm{v}}$ in terms of $L_{\mathrm{v}}^{\mathrm{q}} l_{\mathrm{v}}^{\mathrm{b}}$ and the sample and minibatch sizes.

Let
- $n_v=$ number of validation examples
- $b=$ minibatch size (except possibly the last one)
- $m=\left\lceil\frac{n_v}{b}\right\rceil=$ number of validation minibatches
- $r=n_v-(m-1) b=$ size of the last minibatch (so $1 \leq r \leq b$ )
- $l_v^b=$ loss (mean) on the last minibatch
- $L_v^q=$ "quick \& dirty" estimate = unweighted average of minibatch-mean losses
- $L_v=$ true validation loss $=$ mean over all validation examples

Step 1: Write down what the two averages mean

If $l_1, \ldots, l_{m-1}$ are the mean losses of the first $m-1$ full minibatches (each of size $b$):

Quick estimate:
$$
L_v^q=\frac{1}{m}\left(\sum_{j=1}^{m-1} l_j+l_v^b\right)
$$
True (sample-weighted) validation loss:
$$
L_v=\frac{1}{n_v}\left(b \sum_{j=1}^{m-1} l_j+r l_v^b\right)
$$

Step 2: Eliminate $\sum_{j=1}^{m-1} l_j$ using $L_v^q$

From the first equation:
$$
\sum_{j=1}^{m-1} l_j=m L_v^q-l_v^b
$$
Plug into $L_v$ :
$$
L_v=\frac{1}{n_v}\left(b\left(m L_v^q-l_v^b\right)+r l_v^b\right)=\frac{b m}{n_v} L_v^q+\frac{r-b}{n_v} l_v^b
$$
Final expression (clean form):
$$
L_v=\frac{b m}{n_v} L_v^q+\frac{r-b}{n_v} l_v^b
$$
where $m=\left\lceil n_v / b\right\rceil$ and $r=n_v-(m-1) b$.

We can also write it without $r$ by noting $r-b=n_v-m b$ :
$$
L_v=\frac{b m}{n_v} L_v^q+\frac{n_v-m b}{n_v} l_v^b
$$

### Exercise 2

Show that the quick and dirty estimate $L_{\mathrm{v}}^{\mathrm{q}}$ is unbiased. That is, show that $E\left[L_{\mathrm{v}}\right]=E\left[L_{\mathrm{v}}^{\mathrm{q}}\right]$. Why would you still want to use $L_{\mathrm{v}}$ instead?

Every minibatch mean has the same expectation, so averaging them is unbiased.

$L_v^q$ is unbiased but typically noisier (higher variance) and not the exact sample-mean objective, so $L_v$ is preferred.

### Exercise 3

Given a multiclass classification loss, denoting by $l\left(y, y^{\prime}\right)$ the penalty of estimating $y^{\prime}$ when we see $y$ and given a probabilty $p(y \mid x)$, formulate the rule for an optimal selection of $y^{\prime}$. Hint: express the expected loss, using $l$ and $p(y \mid x)$.

Let the label set be $\mathcal{Y}=\{1, \ldots, K\}$. We observe $x$, and we know the conditional distribution $p(y \mid x)$.

We must choose a prediction $y^{\prime} \in \mathcal{Y}$. The loss is $l\left(y, y^{\prime}\right)$ (true label $y$, predicted label $y^{\prime}$ ).

Step 1: Write the conditional expected loss ("risk") of choosing $y^{\prime}$
Given $x$, the true label $y$ is random with distribution $p(\cdot \mid x)$. If you predict $y^{\prime}$, the expected loss is
$$
\mathcal{L}\left(y^{\prime} \mid x\right)=\mathbb{E}\left[l\left(y, y^{\prime}\right) \mid x\right]=\sum_{y \in \mathcal{Y}} l\left(y, y^{\prime}\right) p(y \mid x) .
$$
Step 2: Choose the $y^{\prime}$ that minimizes that expected loss

So the optimal decision rule is
$$
y^*(x) \in \arg \min _{y^{\prime} \in \mathcal{Y}} \sum_{y \in \mathcal{Y}} l\left(y, y^{\prime}\right) p(y \mid x)
$$
(ties can be broken arbitrarily).
That's the general Bayes decision rule for multiclass classification under arbitrary loss.

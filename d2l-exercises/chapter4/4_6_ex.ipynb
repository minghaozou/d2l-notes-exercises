### Exercise 1

If we wish to estimate the error of a fixed model $f$ to within 0.0001 with probability greater than $99.9 \%$, how many samples do we need?

Apply the Hoeffding Inequality, we get we need roughly $3.45 \times 10^8$ many datapoints

### Exercise 2

Suppose that somebody else possesses a labeled test set $\mathcal{D}$ and only makes available the unlabeled inputs (features). Now suppose that you can only access the test set labels by running a model $f$ (with no restrictions placed on the model class) on each of the unlabeled inputs and receiving the corresponding error $\epsilon_{\mathcal{D}}(f)$. How many models would you need to evaluate before you leak the entire test set and thus could appear to have error 0 , regardless of your true error?

Assume the test set has $n$ examples:

$$
\mathcal{D}=\left\{\left(x^{(i)}, y^{(i)}\right)\right\}_{i=1}^n
$$


You are given only the inputs $x^{(1)}, \ldots, x^{(n)}$, not the labels.
For any model $f$, you can submit predictions on all $n$ inputs, and the oracle returns

$$
\epsilon_{\mathcal{D}}(f)=\frac{1}{n} \sum_{i=1}^n 1\left(f\left(x^{(i)}\right) \neq y^{(i)}\right)
$$


So from one query, you learn only the fraction of mistakes.
The question is:
How many such queries are enough to determine all labels $y^{(i)}$ ?

#### Step 1: Rephrase in terms of unknown bits

To make the logic clean, suppose this is a binary classification problem, so each label is either 0 or 1.
Then the unknown test labels form an $n$-bit string:

$$
\left(y^{(1)}, y^{(2)}, \ldots, y^{(n)}\right)
$$


There are

$$
2^n
$$

possible labelings.
Your goal is to identify which one is the true one.

#### Step 2. What does one model evaluation tell us?

If you submit some classifier $f$, then you get back its error rate.
Since there are $n$ points, the number of mistakes can only be

$$
0,1,2, \ldots, n
$$


So the error rate can only take one of

$$
n+1
$$

possible values:

$$
0, \frac{1}{n}, \frac{2}{n}, \ldots, \frac{n}{n}
$$


That means one query gives at most $n+1$ possible outcomes.
So one query cannot distinguish among more than $n+1$ cases.

#### Step 3. After k queries, how many possibilities can we distinguish

If each query has at most $n+1$ possible answers, then $k$ queries can produce at most

$$
(n+1)^k
$$

different answer patterns.
To uniquely determine the true labeling among all $2^n$ possibilities, we need

$$
(n+1)^k \geq 2^n
$$


Take logs:

$$
k \geq \frac{n \log 2}{\log (n+1)}
$$


So information-theoretically, the number of queries needed is on the order of

$$
\frac{n}{\log n}
$$


This is the minimum scale you should expect.

#### 4. Constructive way to recover all labels

Here is one explicit strategy.
Assume binary labels 0/1.
First submit a classifier $f_0$ that predicts 0 on every test example.
Suppose its error count is $m_0$. Then:
- every point whose true label is 1 is a mistake
- every point whose true label is 0 is correct

So $m_0$ tells you exactly how many 1s are in the test set.
Now for each $j=1, \ldots, n$, submit a classifier $f_j$ that is the same as $f_0$, except on point $x^{(j)}$, where it predicts 1 instead of 0.

Compare the new error count to $m_0$ :
- If the error decreases by 1 , then the true label at point $j$ must be 1
- If the error increases by 1 , then the true label at point $j$ must be 0

So each extra query reveals one label exactly.
Thus:
- 1 query for the all-zero baseline
- $n$ more queries to inspect each point
Total:

$$
n+1
$$


If you already know the baseline count another way, then effectively it is $n$.
So the exercise answer is usually stated as

$$
n \text { (or } n+1 \text { with the baseline query) }
$$

### Exercise 3

What is the VC dimension of the class of fifth-order polynomials?

#### 1. What class of functions are we talking about?

A fifth-order polynomial in one variable has the form

$$
p(x)=a_0+a_1 x+a_2 x^2+a_3 x^3+a_4 x^4+a_5 x^5
$$


So there are 6 coefficients:

$$
a_0, a_1, \ldots, a_5
$$


Usually in VC-dimension questions, we mean the class of binary classifiers obtained by thresholding the polynomial, e.g.

$$
f(x)=1(p(x) \geq 0)
$$


So the label depends on the sign of the polynomial.

#### 2. General rule

For polynomials of degree $d$ on the real line, the VC dimension is

$$
d+1
$$


So for degree 5,

$$
5+1=6
$$


But let's understand why instead of just memorizing it.

#### 3. Why it is at least 6

To show VC dimension is at least 6, we need to show:
There exist 6 points that can be shattered.

Take 6 distinct real numbers

$$
x_1<x_2<\cdots<x_6
$$


Now consider any binary labeling of these 6 points.
A degree-5 polynomial has at most 5 real roots, which means its sign can switch at most 5 times. Across 6 ordered points, that is enough flexibility to realize any pattern of plus/minus signs.

Another way to say it:
- with 6 coefficients, you can interpolate function values at 6 points
- then choose those values to be positive where label is 1 and negative where label is 0
- the interpolating polynomial has degree at most 5

So any labeling of 6 points can be matched.
Hence,

$$
\text { VCdim } \geq 6
$$

#### 4. Why it is not 7

Now take 7 ordered points

$$
x_1<x_2<\cdots<x_7
$$


Consider the alternating labeling

$$
+,-,+,-,+,-,+
$$

or equivalently

$$
1,0,1,0,1,0,1
$$


To realize this pattern, the polynomial's sign must flip between every consecutive pair of points.
That requires at least 6 sign changes, so at least 6 real roots.
But a degree-5 polynomial can have at most 5 real roots.
So this labeling cannot be realized.
Hence 7 points cannot be shattered, and

$$
\text { VCdim } \leq 6
$$

### Exercise 4

What is the VC dimension of axis-aligned rectangles on two-dimensional data?

#### 1. What is the hypothesis class?

An axis-aligned rectangle in 2D looks like

$$
[a, b] \times[c, d]
$$

with sides parallel to the coordinate axes.

As a classifier, it usually means:
- label 1 if the point lies inside the rectangle
- label 0 if the point lies outside

So the question is:
What is the largest number of points in the plane that can be shattered by such rectangles?

#### 2. Show that the VC dimension is at least 4

A standard choice is 4 points arranged as:
- leftmost
- rightmost
- topmost
- bottommost

For example, think of points like

$$
(-1,0),(1,0),(0,1),(0,-1)
$$


These are four points in a cross shape.
We want to argue that every possible labeling of these $\mathbf{4}$ points can be realized by some axis-aligned rectangle.

Why is that possible?
Because an axis-aligned rectangle can always be chosen to include exactly any selected subset of these 4 points:
- one point only: use a tiny rectangle around that point
- two adjacent/extreme points: choose a rectangle spanning just those
- three points: choose a rectangle large enough to include those three but not the excluded extreme point
- all four: choose a large rectangle around all of them
- none: choose a rectangle away from all points
So these 4 points can be shattered.
Thus,

$$
\text { VCdim } \geq 4
$$

#### 3. Show that the VC dimension is at most 4

Take any 5 points in the plane.
Among these 5 points, look at the:
- leftmost point
- rightmost point
- topmost point
- bottommost point

These are at most 4 points total.
Since there are 5 points, at least one point is not one of these four extremes.
So at least one point lies in the "middle," inside the bounding box formed by the extreme points.
Now consider the labeling where:
- the 4 extreme points are labeled 1
- the interior point is labeled 0

Can an axis-aligned rectangle realize this?

No.
Because any axis-aligned rectangle containing the leftmost, rightmost, topmost, and bottommost points must also contain everything in between - in particular, that interior point.

So that labeling is impossible.
Thus no 5-point set can be shattered.

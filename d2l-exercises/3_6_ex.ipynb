When can you solve the problem of polynomial regression exactly?

* d + 1 number of parameters
* n many training points
We can interpolate the data exactly if and only if 
$$d + 1 \geq n$$

### Exercise 3

Can you ever expect to see zero training error? Under which circumstances would you see zero generalization error?

The zero training error occurs when the model capacity is large enough to interpolate the training data. 

The zero generalization error occurs when the true data-generating function lies within the model class and there is no noise (or the noise is predictable).

### Exercise 4

Why is $K$-fold cross-validation very expensive to compute?

Because:
We must retrain the model from scratch $K$ separate times.

If:
- Training once costs $C(N)$,
- Then $K$-fold CV costs approximately:
$$
K \cdot C\left(\frac{K-1}{K} N\right)
$$
Which is roughly:
$$
K \cdot C(N)
$$
So the computational cost increases by a factor of $K$.

### Exercise 5

Why is the $K$-fold cross-validation error estimate biased?

The CV error is biased downward when used for model selection because we choose the model that minimizes the CV estimate, which overfits the validation folds.

### Exercise 6

The VC dimension is defined as the maximum number of points that can be classified with arbitrary labels $\{ \pm 1\}$ by a function of a class of functions. Why might this not be a good idea for measuring how complex the class of functions is? Hint: consider the magnitude of the functions.

Two hypothesis classes can have the same VC dimension (same shattering ability) but behave very differently in terms of complexity/generalization because one class can produce arbitrarily large magnitudes.

Concrete way to see it
Take a real-valued function class $\mathcal{F}$ and turn it into classifiers by thresholding at 0:
$$
h_f(x)=\operatorname{sign}(f(x)) .
$$
Now define a "rescaled" class
$$
\mathcal{F}^{\prime}=\{\alpha f: f \in \mathcal{F}, \alpha>0\} .
$$
Then for any $x$,
$$
\operatorname{sign}(\alpha f(x))=\operatorname{sign}(f(x)) \quad(\alpha>0),
$$
so $\mathcal{F}$ and $\mathcal{F}^{\prime}$ induce exactly the same set of labelings on any finite set of points. Therefore they have the same VC dimension.

But intuitively $\mathcal{F}^{\prime}$ is "bigger" in a meaningful way: it allows arbitrarily large outputs and hence arbitrarily large margins/loss sensitivity, etc. VC dimension won't register any of that, because it discards magnitude information.

Why magnitude matters for "complexity"

Many learning problems don't just use 0-1 loss; they use losses that depend on the real-valued score, e.g.

If $|f(x)|$ can be made huge, we can:
* make training loss tiny (or explode it) without changing classification,
* create unstable fits,
* change generalization properties dramatically.

So a measure of complexity that ignores magnitude can be misleading.

A classic example to keep in mind (linear separators)

In $\mathbb{R}^d$, the VC dimension of hyperplanes is $d+1$, regardless of how large the weight vector $\|w\|$ is. But generalization bounds for margin-based learning depend on things like $\|w\|$ or the margin $\gamma$ (and lead to measures like Rademacher complexity, fat-shattering dimension, margin bounds). VC dimension alone misses that.
